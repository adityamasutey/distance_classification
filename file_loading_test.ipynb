```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Test 1: Check if the dataset file exists
dataset_path = "dataset.csv"  # Replace with your actual dataset filename
if os.path.exists(dataset_path):
    df = pd.read_csv(dataset_path)
    print("Dataset loaded successfully!")
    print(df.head())
else:
    print(f"Dataset file '{dataset_path}' not found!")

# Test 2: Check if images exist and can be loaded
image_path = "harry_potter.png"  # Replace with your actual image filename
if os.path.exists(image_path):
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title("Sample Image")
    plt.show()
    print("Image loaded successfully!")
else:
    print(f"Image file '{image_path}' not found!")
```
